```python
Q1 = data_train['y'].quantile(0.25)
Q3 = data_train['y'].quantile(0.75)
```

```python
IQR = Q3 -Q1
```

```python
lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR
```

```python
print(lower_whisker)
print(upper_whisker)
```

```python
outlier_index = data_train[data_train.y > upper_whisker ].index
outlier_index
```

```python
#drop outlier data 
data_train = data_train.drop(outlier_index)
data_train.info()
```

```python
y = data_train['y']
X = data_train.drop(['y','ID'], axis = 1)
# drop ID column from test data
X_test = data_test.drop(['ID'], axis = 1) 
```
